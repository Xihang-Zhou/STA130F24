```python
# Q8
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split

# Ensure categorical variables have consistent categories
pokeaman["Type 1"] = pd.Categorical(pokeaman["Type 1"], 
                                         categories=pokeaman["Type 1"].unique())
pokeaman["Type 2"] = pd.Categorical(pokeaman["Type 2"], 
                                         categories=pokeaman["Type 2"].unique())
pokeaman["Generation"] = pd.Categorical(pokeaman["Generation"])

# Define the number of repetitions
reps = 100

# Define model specifications for models 3, 4, 5, 6, and 7
model_specs = {
    "Model 3": 'HP ~ Attack + Defense + Speed',
    "Model 4": 'HP ~ Attack * Defense * Speed * Legendary * Q("Sp. Def") * Q("Sp. Atk")',
    "Model 5": 'HP ~ Attack + Defense + Speed + Legendary + Q("Sp. Def") + Q("Sp. Atk") + C(Generation) + C(Q("Type 1")) + C(Q("Type 2"))',
    "Model 6": 'HP ~ Attack + Speed + Q("Sp. Def") + Q("Sp. Atk") + I(Q("Type 1")=="Normal") + I(Q("Type 1")=="Water") + I(Generation==2) + I(Generation==5)',
    "Model 7": 'HP ~ Attack * Speed * Q("Sp. Def") * Q("Sp. Atk") + I(Q("Type 1")=="Normal") + I(Q("Type 1")=="Water") + I(Generation==2) + I(Generation==5)'
}

# Initialize a DataFrame to store results
results = pd.DataFrame(columns=["Model", "In Sample R-squared", "Out of Sample R-squared"])

# Loop over each model specification
for model_name, linear_form in model_specs.items():
    in_sample_Rsquared = np.zeros(reps)
    out_of_sample_Rsquared = np.zeros(reps)
    
    # Perform 100 repetitions with 50-50 train-test splits
    for i in range(reps):
        # 50-50 train-test split
        pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=0.5)
        
        # Ensure consistent categories in train and test sets
        pokeaman_train["Type 1"] = pokeaman_train["Type 1"].cat.set_categories(pokeaman["Type 1"].cat.categories)
        pokeaman_train["Type 2"] = pokeaman_train["Type 2"].cat.set_categories(pokeaman["Type 2"].cat.categories)
        pokeaman_test["Type 1"] = pokeaman_test["Type 1"].cat.set_categories(pokeaman["Type 1"].cat.categories)
        pokeaman_test["Type 2"] = pokeaman_test["Type 2"].cat.set_categories(pokeaman["Type 2"].cat.categories)

        # Fit the model
        model_fit = smf.ols(formula=linear_form, data=pokeaman_train).fit()
        
        # Collect in-sample R-squared
        in_sample_Rsquared[i] = model_fit.rsquared
        
        # Collect out-of-sample R-squared
        y_test = pokeaman_test.HP
        y_pred_test = model_fit.predict(pokeaman_test)
        out_of_sample_Rsquared[i] = np.corrcoef(y_test, y_pred_test)[0, 1] ** 2
    
    # Append results to DataFrame
    model_results = pd.DataFrame({
        "Model": model_name,
        "In Sample R-squared": in_sample_Rsquared,
        "Out of Sample R-squared": out_of_sample_Rsquared
    })
    results = pd.concat([results, model_results], ignore_index=True)

# Visualize the results
fig = px.scatter(results, x="In Sample R-squared", y="Out of Sample R-squared",
                 color="Model", title="In Sample vs Out of Sample R-squared across Different Models")

# Add a y=x line for reference
fig.add_trace(go.Scatter(x=[0, 1], y=[0, 1], name="y=x", line=dict(dash="dash")))

# Show the plot
fig.show(renderer='png')
```

This graph shows generalizability and correctness of the model, the higher the better result. If x is far larger than y, then over fitting occurs, model 6 seems to show the best result.
